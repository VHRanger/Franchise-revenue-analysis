```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
sns.set_palette("muted")
```

```python
df = pd.read_csv('franchises.csv')
df['Retail & Licensing'] = df.Merchandise + df['Retail (unattributed or licensing)'] + df.Museum
df = df.drop(['Merchandise', 'Retail (unattributed or licensing)'], 1)
df['TV & Home Video'] = df.TV + df['Home Entertainement (w/ TV)']
df = df.drop(['TV', 'Home Entertainement (w/ TV)'], 1)

df = df.set_index('Franchise')

df['total'] = df.sum(axis=1)
df = df.sort_values(by='total', ascending=True)
df = df.drop('total', 1)

# Drop small values
df['Other'] = df.Music + df['Stage Play'] + df.Museum
df = df.drop(['Music', 'Stage Play', 'Museum'], 1)
```

```python
df = df[['Retail & Licensing', 'Video Games', 'Box Office',
         'TV & Home Video', 'Books/Mangas', 'Card Games', 
         'Pachinko', 'Other']]

df
```

```python
df.plot(kind='barh', figsize=(13,13), stacked=True)
plt.xlabel('$ (millions)')
plt.show()
```
